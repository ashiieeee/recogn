```python
import numpy as np
import math

# Function to calculate mean
def calculate_mean(values):
    return np.mean(values)

# Function to calculate maximum
def calculate_max(values):
    return np.max(values)

# Function to calculate minimum
def calculate_min(values):
    return np.min(values)

# Function to calculate energy
def calculate_energy(values):
    return math.sqrt(np.sum(np.square(values)))

# Function to calculate entropy
def calculate_entropy(values):
    unique_values, counts = np.unique(values, return_counts=True)
    probabilities = counts / len(values)
    entropy = -np.sum(probabilities * np.log2(probabilities))
    return entropy

# Function to calculate angle
def calculate_angle(mean_value, axis_value):
    cosine_angle = np.dot(mean_value, axis_value) / (np.linalg.norm(mean_value) * np.linalg.norm(axis_value))
    angle = np.arccos(cosine_angle)
    return np.degrees(angle)

# Input values
x_values = [float(input("Enter x value: ")) for _ in range(4)]
y_values = [float(input("Enter y value: ")) for _ in range(4)]
z_values = [float(input("Enter z value: ")) for _ in range(4)]
gravity_mean = np.array([0, -9.81, 0])  # Assuming gravity along Y axis

# Calculate features
mean_x = calculate_mean(x_values)
mean_y = calculate_mean(y_values)
max_x = calculate_max(x_values)
max_y = calculate_max(y_values)
min_x = calculate_min(x_values)
min_y = calculate_min(y_values)
energy_z = calculate_energy(z_values)
entropy_x = calculate_entropy(x_values)
entropy_y = calculate_entropy(y_values)
angle_with_gravity = calculate_angle(gravity_mean, np.array([0, 1, 0]))

# Print results
print(f"tGravityAcc-mean()-X: {mean_x}")
print(f"tGravityAcc-mean()-Y: {mean_y}")
print(f"tGravityAcc-max()-X: {max_x}")
print(f"tGravityAcc-max()-Y: {max_y}")
print(f"tGravityAcc-min()-X: {min_x}")
print(f"tGravityAcc-min()-Y: {min_y}")
print(f"tGravityAcc-energy()-Z: {energy_z}")
print(f"tGravityAcc-entropy()-X: {entropy_x}")
print(f"tGravityAcc-entropy()-Y: {entropy_y}")
print(f"angle(Y,gravityMean): {angle_with_gravity}")
```
